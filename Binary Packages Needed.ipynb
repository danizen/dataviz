This notebook determines the binary packages needed for Windows and MacOS to assure parity with the union of nlm-occs-devops-py36-el7 and nlm-occs-devops-py38-el7.  The input CSVs were produced using the `findbinary.py` script which produces a CSV of all binary packages via the PyPI simple API against a repository on our server.

So, 

```bash
./findbinary.py nlm-occs-devops-py36-el7 > binary-py36.csv
```

Will go to https://tools.nlm.nih.gov/artifactory/api/pypi/nlm-occs-devops-py36-el7/simple/ and crawl to find all binary versions and build a CSV.

````python
import pandas as pd
````

````python
df36 = pd.read_csv('binary-py36.csv')
df36
````

````python
df38 = pd.read_csv('binary-py36.csv')
df38
````

````python
df36 = df36.drop('Repository Name', axis=1)
df36
````

````python
df38 = df38.drop('Repository Name', axis=1)
df38
````

````python
df = pd.concat([df36, df38])
df
````

````python
df = df.sort_values(['Package', 'Version'])
df
````

````python
all_binary = df.drop_duplicates(subset=['Package', 'Version'], keep='last').reset_index(drop=True)
all_binary
````

````python
all_binary = all_binary.drop('File', axis=1)
all_binary
````

````python
all_binary.to_csv('binary.csv', index=False)
````
