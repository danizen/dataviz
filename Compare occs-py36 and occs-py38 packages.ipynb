```python
import pandas as pd
```

```python
df36 = pd.read_csv('occs-py36.csv', index_col='Package')
df36
```

```python
df38 = pd.read_csv('occs-py38.csv', index_col='Package')
df38
```

```python
joined = df36.join(df38, how='outer', lsuffix='_36', rsuffix='_38')
joined
```

```python
missing_36 = joined[joined['nlm-occs-devops-py36-el7'].isnull()]
missing_36
```

```python
missing_38 = joined[joined['nlm-occs-devops-py38-el7'].isnull()]
missing_38
```

```python
notnull = (joined['nlm-occs-devops-py36-el7'].notnull()) & (joined['nlm-occs-devops-py38-el7'].notnull())
notnull
```

```python
common = joined[notnull]
common
```

```python
mismatch = common[common['nlm-occs-devops-py36-el7'] != common['nlm-occs-devops-py38-el7']]
mismatch
```

```python
with open('mismatch.txt', 'w') as f:
    for package in mismatch.index:
        print(str(package), file=f)
```

```python
new_col = (mismatch['Can Upgrade_36'] == 'Yes')|(mismatch['Can Upgrade_38'] =b= 'Yes')
new_col
```

```python
mismatch = pd.DataFrame({
    'py36_version': mismatch['nlm-occs-devops-py36-el7'],
    'py38_version': mismatch['nlm-occs-devops-py38-el7'],
    'can_upgrade': new_col,
}, index=mismatch.index)
mismatch
```

```python
mismatch.to_csv('mismatch.csv')
```
