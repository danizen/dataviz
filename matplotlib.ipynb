The "figsize" takes a tuple or list that controls the figure size in inches.

```python
fig = plt.figure(figsize=(13, 9))
```

You cannot plot just on a figure, you need to create sub-plots to do this.

```python
plt.plot(np.arange(10))
plt.plot(np.random.randn(10))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
# histogram
axes[0,0].hist(np.random.randn(10))
# scatter
axes[0,1].scatter(np.arange(30), np.arange(30) + (3 * np.random.randn(30)))
# line (with style)
axes[1,0].plot(np.random.randn(50).cumsum(), 'k--')
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(13, 8))
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, alpha=0.5)
plt.subplots_adjust(hspace=0, wspace=0)
```

```python
plt.figure(figsize=(13, 8))
plt.plot(np.random.randn(30).cumsum(), 'g^--')
```

Different ways of interpolating between points

```python
data = np.random.randn(30).cumsum()
fig = plt.figure(figsize=(13, 8))
plt.plot(data, 'b--', label='Default')
plt.plot(data, 'g-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

### Ticks, Labels, and Legends

Procedural and object oriented methods corresponding to plt. procedural methods on latest subplot (axes) or object oriented methods on the subplot.

* `plt.xlim()` means `ax.get_xlim()`
* `plt.xlim(*args, **kwargs)` means `ax.set_xlim(*args, **kwargs)`.

See also:

* `plt.xticks?`
* `plt.ylim?`
* `plt.yticks?`

Big example plot follow.

```python
dataset = np.random.randn(1000).cumsum()
```

```python
fig = plt.figure(figsize=(13, 8))
ax = fig.add_subplot(1, 1, 1)
ax.plot(dataset)

# setting ticks, ticklabels, and xlabel via methods
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(
    ['one', 'two', 'three', 'four', 'five'],
    rotation=30,
    fontsize='small'
)
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(13, 8))
ax.plot(dataset)

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(
    ['A', 'B', 'C', 'D', 'E'],
    rotation=90
)

# we can also set these labels via property settings
props = {
    'title': 'My second matplotlib plot',
    'xlabel': 'Stages',
}
ax.set(**props)
```

#### Adding Legends
When multiple datasets are plotted on the same AxesSubplot, they can each have labels.  These are then displayed by calliing something like `ax.legend(loc='best')`.

```python
ds1 = np.random.randn(1000).cumsum()
ds2 = np.random.randn(1000).cumsum()
ds3 = np.random.randn(1000).cumsum()
```

```python
fig = plt.figure(figsize=(13, 8))
ax = fig.add_subplot(1, 1, 1)

ax.plot(ds1, 'b', label='one')
ax.plot(ds2, 'g--', label='two')
ax.plot(ds3, 'c-.', label='three')

ax.legend(loc='best')
```
